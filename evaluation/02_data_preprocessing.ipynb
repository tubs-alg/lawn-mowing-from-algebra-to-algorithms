* Instance information
    * `polygon` Polygon. Read with `shapely.wkt.loads`
    * `polygon_area` Area of polygon
    * `convex_hull_area` Area within the convex hull of the polygon
    * `relative_area` Area within convex hull relative to tool size
    * `instance_path` Path to the source of this instance
    * `instance` Name of this instance
    * `instance_type` Name of the instance generator
        * `instance_type_simplified` Simplified instance type description (less cryptic)
    * `grid_properties` Some data on the sparse grid
* **Dense grid in which visiting every vertex suffices**
    * TSP on dense grid in which visiting every vertex suffices for full coverage
        * `small_tsp_path` The path of the TSP in the dense grid ` `[[x1, y1], [x2,y2],..., [x1, y1]]``
        * `small_tsp_obj` The length of above's tour
        * `small_tsp_runtime` The runtime in seconds to compute it
* **Sparse grid that requires modification for full coverage**
    * TSP on sparse grid *before* modification for full coverage
        * `timelimit_tsp` The time limit (s) for this step.
        * `tsp_path_grid_graph` The path of the TSP in the grid, before applying the modification for full coverage. ` `[[x1, y1], [x2,y2],..., [x1, y1]]``
        * `tsp_obj_grid_graph` The length of above's tour
    * TSP *after* modification for full coverage.
        * `tsp_path` The path of the modified TSP tour. ` `[[x1, y1], [x2,y2],..., [x1, y1]]``
        * `tsp_obj` Its length.
        * `tsp_runtime` The runtime in seconds to compute it
        * (`tsp_path_before_optimization` The path of the modified TSP tour if we just make any passage through a tile covering. ` `[[x1, y1], [x2,y2],..., [x1, y1]]``)
    * After optimizing the previous tour using an LNS that is sensitive to the modification costs.
        * `timelimit_lns` The time limit (s) for this step.
        * `lns_path` The path of this tour. `[[x1, y1], [x2,y2],..., [x1, y1]]`
        * `lns_obj` Its length
        * `lns_runtime` The runtime of this optimization.
        * `lns_is_optimal` If this tour is provable optimal (within the used grid, not geometrically)
        * (`lns_path_before_optimization` path before performing some local optimization)
    * After solving a MIP that can potentially compute an optimal solution within the specific grid. It starts with the previous path (LNS).
        * `timelimit_mip` The time limit (s) for this step.
        * `mip_path` The path of this tour. `[[x1, y1], [x2,y2],..., [x1, y1]]`
        * `mip_obj` Its length
        * `mip_runtime` The runtime of this optimization.
        * `lb` The lower bound computed by the MIP
        * `ub` The upper bound computed by the MIP. Should be identical to `mip_obj`.
        * (`mip_path_before_optimization` path before performing some local optimization)
